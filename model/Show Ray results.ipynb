```python
import os
import datetime
import json
base_dir = '/home/dasolma/ray_results/train_2023-04-05_09-48-14'
data = []
for exp in os.listdir(base_dir):
    if 'json' in exp or 'pkl' in exp:
        continue
    file = os.path.join(base_dir, exp, 'result.json')
    if os.path.exists(file):
        print(datetime.datetime.fromtimestamp(os.path.getmtime(file)))
        try:
            r = json.load(open(file, 'r'))
            print(r['score'], r[ 'time_this_iter_s'] / 3600, exp[:14])
            data.append((r['score'], r[ 'time_this_iter_s'] / 3600, exp[:14], os.path.getmtime(file)))
        except:
            print(exp, "error")
            pass
    
```

```python
data = sorted(data, key=lambda x: -x[0])
```
